```python
import pandas as pd
import os

total_dollars_avail = 200 * 12 # 200 dollars per team, 12 teams

raw_rankings_folder = "raw_rankings"
ranked_folder = "valued_ranks"

for csv_file in os.listdir(raw_rankings_folder):
    if csv_file.endswith(".csv"):
        file_name = csv_file.split(".")[0]
        data = pd.read_csv(f"{raw_rankings_folder}/{csv_file}")
        # convert to numeric
        data["TOTAL"] = pd.to_numeric(data["TOTAL"], errors="coerce")
        # delete rows with nan valeus
        data = data.dropna(subset=["TOTAL"])
        # get total Z scores
        total_z = data[data["TOTAL"] > 0]["TOTAL"].sum()
        print(f"total Z scores all +ve players in {file_name}: ", total_z)
        
        data["VALUE"] = ((data["TOTAL"] / total_z) * total_dollars_avail).round().apply(lambda x: f"${x}")

        # Keep only the specified columns
        filtered_data = data[["R#", "ADP","PLAYER", "VALUE"]]
        # if rank cell has space in between, split it and keep the first part
        filtered_data.loc[:, "R#"] = filtered_data["R#"].str.split().str[0]

        # Convert the DataFrame to a CSV file
        filtered_data.to_csv(f"{ranked_folder}/{file_name}-ranked.csv", index=False)
```
