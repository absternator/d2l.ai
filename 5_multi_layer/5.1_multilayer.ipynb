### Multilayer perceptrons 
- linearity implies weak assumption of monotonicity(if var increase, output increase)
- we can overcome issues by adding hidden layers 
- nonlinear activation function $\sigma$ applied to hidden layers -- eg. ReLU, tanh, sigmoid
$$
\begin{aligned}
    \mathbf{H} & = \sigma(\mathbf{X} \mathbf{W}^{(1)} + \mathbf{b}^{(1)}), \\
    \mathbf{O} & = \mathbf{H}\mathbf{W}^{(2)} + \mathbf{b}^{(2)}.\\
\end{aligned}
$$
- we can continue stacking more hidden layers, yielding even more expressive models
- activation functions decide wether a neuron should be activated or not.

```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

```python
# Relu = Relu(x) = max(x, 0) ... anything negative is 0 and turns off
x = torch.arange(-8, 8, 0.1, requires_grad=True)
y = torch.relu(x)
d2l.plot(x.detach(), y.detach(), "x", "relu(x)")
```

```python
# derivative of relu.. 0 for -ve values and 1 for positive values 
y.sum().backward(retain_graph=True) #y.backward(torch.ones_like(x), retain_graph=True)
d2l.plot(x.detach(), x.grad, "x", "relu grad")
# note there are variants of relu
```

```python
# note there are variants of relu.. adds alpha term so give some info to negative
x = torch.arange(-8, 8, 0.1, requires_grad=True)
y = torch.nn.PReLU()
d2l.plot(x.detach(), y(x).detach(), "x", "prelu(x)")
```

```python
# sigmoid 1/(1+e^-x)- squashing func... any input (-inf, inf) to (-1, 1)
# binary classification (special case of softmax) ... not used as much Relu used more as sigmoid has issues 
y = torch.sigmoid(x)
d2l.plot(x.detach(), y.detach(), "x", "sigmoid(x)", figsize=(5, 2.5))
```

```python
y.sum().backward(retain_graph=True)
d2l.plot(x.detach(), x.grad, "x", "derivate sigmoid", figsize=(5, 2.5))
```

```python
# tanh - transords to (-1, 1) again = (1 - e^-2x)/(1 + e^-2x) -- symetry about centre
y = torch.tanh(x)
d2l.plot(x.detach(), y.detach(), "x", "tanhx(x)")
```

```python
y.sum().backward(retain_graph=True)
d2l.plot(x.detach(), x.grad, "x", "derivate tanh", figsize=(5, 2.5))
```
