# Intro 
- predictions can be distributions, not just point estimates
- kernel $K$ a covariance function, $K(x,x')$ is the covariance between $x$ and $x'$
- gaussian process is a distribution over functions, not a distribution over parameters
  - instead of conventional correlation matrix, we have a covariance function that generalixe to infinite dimensions/observations/predictions
  - kernel gives distance b/w points, and how correlated they are (distance can be anything(time, space, etc)). Continuous ordered categories
-  
-  kernel function determines how "wigily" curves/ how free they  can be b/w points
-  we actually learn covariance kernel function from data... kinda like CNNs learn filters                                                                                        
-  some choices of kernel functions: quadratic(L2), periodic, ornestein-uhlenbeck, radial basis function(RBF), etc
-  specify hyperparams using marginal likelihood
