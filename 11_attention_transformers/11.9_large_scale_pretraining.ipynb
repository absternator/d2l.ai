# 11.9. Large-Scale Pretraining with Transformers
- Really good scaling behaviour
- 3 modes: encoder-only, decoder-only, encoder-decoder
- Encoder-only: Vision Transformers, BERT(masked language model), bidirectional (depends on input tokens before and after). can be fine-tuned further!!
- Decoder-only: GPT-2, GPT-3 - all better without changing architecture, parameter updates. Unidirectional (only depends on tokens before). Can be fine-tuned further!!
  - GPT-2: 1.5B parameters, 40GB of text, 40% of the internet, 1 month of training on 256 GPUs
  - GPT-3: 175B parameters, 570GB of text, 300B tokens, 1 year of training on 1024 GPUs
  - few-shot much better 
- Encoder-decoder: Need to train arbitrary lengths. BART & T5 (text-to-text transfer transformer), can do like summarize, translate, etc. can also be fine-tuned.

- Scalability 
  - follows a power law
